```python
import random
import matplotlib.pyplot as plt
import timeit
from timeit import default_timer as timer

def gen_random(size):
        x = []
        for m in range(size):
            v = random.randint(0,9999999)
            x.append(v)
        return x

def gen_increasing(size):
    x = []
    for m in range(size):
         v = random.randint(0,99)
         x.append(v)
         x.sort()
    return x

def gen_decreasing(size):
    x = []
    for m in range(size):
         v = random.randint(0,99)
         x.append(v)
         x.sort()
         x.reverse()
    return x
        
def ins(a, n):
    start = timer()
    for i in range(1, n):
        k = a[i]
        j = i - 1

        while a[j]>k and j >= 0:
            a[j + 1] = a[j]
            j = j - 1
        a[j + 1] = k
        
    stop = timer()
    te = stop - start
    return te, a
   
sizes = [5,10,50,100,250,500,1000,2000,5000,10000,20000]
jr = []
ji = []
jd = []
for c in sizes:
    print(f'running size {c}')
    s = gen_random(c)
    ss = gen_increasing(c)
    sss = gen_decreasing(c)
    t, z = ins(s, c)
    tt, z = ins(ss, c)
    ttt, z = ins(sss, c)
    jr.append(t)
    ji.append(tt)
    jd.append(ttt)
print('1',jr,'2',ji,'3',jd)

    

    
    
    
        
            
            
    
    
    
```
