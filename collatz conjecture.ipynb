```python
import matplotlib.pyplot as plt
def col(x):
    y = []
    count = 0
    while x != 1:
        count += 1
        if x%2 == 0:
            x = x // 2
            y.append(x)
        else:
            x = 3*x + 1
            y.append(x)
    #plt.plot(y)
    #plt.scatter(y,count)
    return count, y
   

#l, path = col(7)
## append l to a list, plt.scatter that list
## call col(i) for i between 2 - 1000
m = [0, 0]
for i in range(2, 10000):
    l, path = col(i)
    m.append(l)
#print(path)
```
